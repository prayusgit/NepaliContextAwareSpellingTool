```python
confusion_set_df = pd.read_excel('./datasets/words_to_create_test_set_no_duplication.xlsx')
```

```python
import pandas as pd

# Sample data as described

# Initialize an empty dictionary
word_dict = {}

# Populate the dictionary
for _, row in confusion_set_df.iloc[:,:2].iterrows():
    word = row['Word_1']
    confusion_words = [word.strip() for word in row['Confusion'].split(',')]  # Handle multiple confusion words
    word_dict[word] = confusion_words

    # Add reverse mappings for all confusion words
    for confusion_word in confusion_words:
        if confusion_word not in word_dict:
            word_dict[confusion_word] = []
        if word not in word_dict[confusion_word]:
            word_dict[confusion_word].append(word)

# Print the resulting dictionary
print(word_dict)
```

```python
import pandas as pd


# Function to remove post-positions
def remove_post_positions(sentence, confusion_set, post_positions):
    words = sentence.split()  # Split sentence into words
    cleaned_words = []

    for i, word in enumerate(words):
        # Check if the word matches a confusion word or its variants
        for confusion_word, variants in confusion_set.items():
            if word.startswith(confusion_word):
                # Remove post-position if it exists
                for post_position in post_positions:
                    if word == confusion_word + post_position:
                        word = confusion_word  # Remove the post-position
                        break
        cleaned_words.append(word)
    
    return " ".join(cleaned_words)

# Apply the function to each row in the dataset
df['text'] = df['text'].apply(lambda x: remove_post_positions(x, word_dict, post_positions))
```
