```python
import json
```

```python
# Loading Lexicon

file_path = 'sabdakosh.json'

# Read the JSON file
with open(file_path, 'r', encoding='utf-8') as file:
    sabdakosh = json.load(file)

# Print the data
print(sabdakosh)
```

```python
# Making sabdakosh list
words = []
for sabda in sabdakosh:
  words.append(sabda['word'])
```

```python
'कटक बिजुली' in words
```

```python
# Grouping the words that started with same letter

from collections import defaultdict


grouped_words = defaultdict(list)
for word in words:
    first_letter = word[0]
    grouped_words[first_letter].append(word)

# Display grouped words
for letter, words in grouped_words.items():
    print(f"Words starting with '{letter}': {words}")
```

```python
%time
from nltk.metrics import edit_distance
import pickle

# Compute confusion sets for each group
confusion_sets = defaultdict(list)
max_edit_distance = 1

for letter, words in grouped_words.items():
    for i, word1 in enumerate(words):
      if len(word1) > 1:
        for word2 in words[i+1:]:
          if edit_distance(word1, word2) <= max_edit_distance:
            confusion_sets[word1].append(word2)
            confusion_sets[word2].append(word1)

# Display confusion sets
# for word, confusions in confusion_sets.items():
#     print(f"Confusions for '{word}': {confusions}")
with open('confusion_sets.json', 'wb') as f:
  pickle.dump(confusion_sets, f)
```
