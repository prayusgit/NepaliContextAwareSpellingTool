```python
def transliterate_first_letter(word):
    first_letter = word[0]
    return transliteration_map.get(first_letter, first_letter)

# Group words by transliteration of the first letter
grouped_words = defaultdict(list)
for word in words_2:
    transliterated_letter = transliterate_first_letter(word)
    grouped_words[transliterated_letter].append(word)

# Display grouped words
for letter, words in grouped_words.items():
    print(f"Words starting with transliterated '{letter}': {words}")
```

```python
from nltk.metrics import edit_distance
import pickle

# Compute confusion sets for each group
confusion_sets = defaultdict(list)
max_edit_distance = 1
ikar_ukar = ['ि', '◌ी', '◌ू','◌ु']

def remove_ikar_ukar(word):
  for sub in ikar_ukar:
    word = word.replace(sub, '')
  return word

for letter, words in grouped_words.items():
    for i, word1 in enumerate(words):
      if len(word1) > 1:
        for word2 in words[i+1:]:
          if len(word2) > 1:
            if edit_distance(remove_ikar_ukar(word1[1:]), remove_ikar_ukar(word2[1:])) <= max_edit_distance:
              confusion_sets[word1].append(word2)
              confusion_sets[word2].append(word1)

# Display confusion sets
# for word, confusions in confusion_sets.items():
#     print(f"Confusions for '{word}': {confusions}")
with open('confusion_sets.json', 'wb') as f:
  pickle.dump(confusion_sets, f)
```
